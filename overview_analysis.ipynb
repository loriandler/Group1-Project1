```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "../Group1-Project1/Output/School_Summary.csv"
data_df = pd.read_csv(file_path)
data_df
```

```python
n=5
top_n_highest_completion_rate_list = data_df['Overall Completion Rate'].nlargest(n=n).to_list()

bottom_n_lowest_completion_rate_list = data_df['Overall Completion Rate'].nsmallest(n=n).to_list()
```

```python
def loc_funtion(x):
    y = data_df.loc[data_df['Overall Completion Rate'] == x]
    return y
```

```python
top_n_df_list = []
for index in range(len(top_n_highest_completion_rate_list)):
    df = loc_funtion(top_n_highest_completion_rate_list[index])
    top_n_df_list.append(df)
    top_n_universities_with_highest_completion_rate_df = pd.concat(top_n_df_list)
```

```python
bottom_n_df_list = []
for index in range(len(bottom_n_lowest_completion_rate_list)):
    df = loc_funtion(bottom_n_lowest_completion_rate_list[index])
    bottom_n_df_list.append(df)
    bottom_n_universities_with_lowest_completion_rate_df = pd.concat(bottom_n_df_list)
```

```python
bottom_n_universities_with_lowest_completion_rate_df = bottom_n_universities_with_lowest_completion_rate_df[['School Name', 'School City', 'Institution Type',
       'Total Students', '% Male Students', '% Female Students',
       '% Caucasian Student', '% African American Student',
       '% Hispanic Student', '% Asian Student',
       '% American Indian and Alaska Native Student',
       '% Native Hawaiian and Pacific Islander Student',
       '% Two or More Race Student', '% Non Resident Alien Student',
       '% Unknown Student', 'Retention Rate',
       'Share of Student Received Pell Grant', 'Overall Completion Rate',
       'Completion rate of Caucasian Student',
       'Completion rate of African American Student',
       'Completion rate of Hispanic Student',
       'Completion rate of Asian Student',
       'Completion rate of American Indian/Alaska Native Student',
       'Completion rate of Two or More Races Student',
       'Completion rate of Non Tesident Alien Student']].reset_index(drop=True)
bottom_n_universities_with_lowest_completion_rate_df
```

```python
top_n_universities_with_highest_completion_rate_df = top_n_universities_with_highest_completion_rate_df[['School Name', 'School City', 'Institution Type',
       'Total Students', '% Male Students', '% Female Students',
       '% Caucasian Student', '% African American Student',
       '% Hispanic Student', '% Asian Student',
       '% American Indian and Alaska Native Student',
       '% Native Hawaiian and Pacific Islander Student',
       '% Two or More Race Student', '% Non Resident Alien Student',
       '% Unknown Student', 'Retention Rate',
       'Share of Student Received Pell Grant', 'Overall Completion Rate',
       'Completion rate of Caucasian Student',
       'Completion rate of African American Student',
       'Completion rate of Hispanic Student',
       'Completion rate of Asian Student',
       'Completion rate of American Indian/Alaska Native Student',
       'Completion rate of Two or More Races Student',
       'Completion rate of Non Tesident Alien Student']].reset_index(drop=True)
top_n_universities_with_highest_completion_rate_df
```

```python
top_n_output_path = f"Output/top_{n}_highest_completion_rate_university.csv"
top_n_universities_with_highest_completion_rate_df.to_csv(top_n_output_path)
```

```python
bottom_n_output_path = f"Output/bottom_{n}_lowest_completion_rate_university.csv"
bottom_n_universities_with_lowest_completion_rate_df.to_csv(bottom_n_output_path)
```
