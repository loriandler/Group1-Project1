```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

school_data_to_load = Path("Output/School_Summary.csv")

school_data = pd.read_csv(school_data_to_load)

school_data
```

```python
school_data.columns
```

```python
institution_types = school_data["Institution Type"].unique().tolist()
institution_types
```

```python
public_data = school_data[school_data["Institution Type"] == "Public"]
public_data
```

```python
public_completion_rate = public_data["Overall Completion Rate"]
public_completion_rate
```

```python
average_public_comp = public_completion_rate.mean()
average_public_comp
```

```python
private_data = school_data[school_data["Institution Type"] == "Private nonprofit"]
private_data
```

```python
private_completion_rate = private_data["Overall Completion Rate"]
private_completion_rate
```

```python
avg_private_comp = private_completion_rate.mean()
avg_private_comp
```

```python
x_axis = ["Public", "Private"]
y_axis = [average_public_comp, avg_private_comp]
```

```python
plt.bar(x_axis, y_axis, color="r", alpha = 0.5, align= "center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Public", "Private"])
plt.xlim(-0.75, 1.75)
plt.ylim(0, max(y_axis) + 0.1)
plt.title("Public vs. Private Average Completion Rates")
plt.xlabel("School Type")
plt.ylabel("Average Completion Rate (decimal)")
plt.savefig("Output/PrivateVsPublicAvg.png")
plt.show()
```
